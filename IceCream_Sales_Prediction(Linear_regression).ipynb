```python
X = IceCream[['Temperature']]
Y = IceCream[['Revenue']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split
```

```python
 X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

### Trainig Model-Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression(fit_intercept = True)
regressor.fit(X_train,Y_train)
```

```python
print('Linear Model Coefficient (m)',regressor.coef_)
print('Linear Model Intercept (b)',regressor.intercept_)
```

### Testing Model-Linear Regression

```python
Y_predict = regressor.predict(X_test)
```

```python
plt.scatter(X_train, Y_train, color='gray')
plt.plot(X_train, regressor.predict(X_train), color='red')
plt.ylabel('Revenue')
plt.xlabel('Temperature')
plt.title('Revenue vs. Temrature on beach(Training set)')
```

```python
plt.scatter(X_test, Y_test, color='gray')
plt.plot(X_test, regressor.predict(X_test), color='green')
plt.ylabel('Revenue')
plt.xlabel('Temperature')
plt.title('Revenue vs. Temrature on beach(Testing set)')
```
